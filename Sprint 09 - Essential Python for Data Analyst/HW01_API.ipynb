```python
from requests import get
import pandas as pd
from time import sleep

url = "https://api.openbrewerydb.org/v1/breweries/"
response = get(url)
data = response.json()

nums = list(range(0,49))
result = []

for i in nums:
    row = [
        data[i]['name'],
        data[i]['brewery_type'],
        data[i]['state'],
        data[i]['website_url']
    ]
    result.append(row)
    sleep(0.5)

print(result)
```

```python
df = pd.DataFrame(result, columns = ['name','brewery type','state','website'])
df
```
